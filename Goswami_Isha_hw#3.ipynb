```python
"""
Defining the simDNA function that calculates the number of substitutions between two DNA
sequences based on the calculation scheme outlined in the homework problem, similar to last
week's.
"""
def simDNA(seq1, seq2):
  if(len(seq1) != len(seq2)):
    print("Seqences must have the same length")
    return
  numNotSame = 0
  for i in range(len(seq1)):
    if(seq1[i] != '-' and seq2[i] != '-'):

      if(seq1[i] != seq2[i]):
        numNotSame +=1

  return numNotSame
```

```python
"""
 Defining the readFile function that reads each of the data files and stores
 the names of the tortoise species and their corresponding DNA sequence in a 2D-list.
 Each element of the list has two elements: the tortoise species and its DNA sequence.
"""
def readFile(filename):
  listFrogs = []

  with open(filename, 'r') as file:
    seq = ""
    name = ""
    for line in file:
      if(line[0] == '>' and seq == ""):
        name = line.strip('\n')[1:]
      elif(line[0] == '>' and name != ""):
        listFrogs.append([name, seq])
        name = line.strip('\n')[1:]
        seq = ""
      else:
        seq += line.strip('\n')

  listFrogs.append([name, seq])
  return listFrogs
```

```python
"""Call the readFile function to store the specie names and DNA sequences
of the turtles from the first file, used for 2b"""
turtles = readFile("alignmentOutput.txt");

k = 1
for j in range(len(turtles)):
    for i in range(j+1, len(turtles)):
     #call the function defined above to find the number of substitutions between each
     #unique pair of DNA sequences
      val = simDNA(turtles[i][1], turtles[j][1]);
      print("The number of substitutions between samples ", turtles[i][0], " and ", turtles[j][0], " is: ", val);
```

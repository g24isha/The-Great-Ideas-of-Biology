```python
import numpy as np
```

```python
"""
The following code reads the file and stores the values in a matrix.
"""
energy_matrix = np.zeros((41, 4))
j = 0
with open("RNAP_matrix.txt", 'r') as file:
    for line in file:
      if(line[0] != '#' and line[0] != 'A'):
        ACGT = line.split()
        for i in range(4):
          num = ACGT[i]
          exponent = num.split('e')
          energy_matrix[j][i] = float(exponent[0]) *(10 ** float(exponent[1]))
        j+=1
```

```python
"""
Define a function that gets the binding energy of a sequence by
iterating through the letters of the sequence and adding the
corresponding energy from the given matrix, to get the final
value of binding energy.
"""
def get_energy(sequence, matrix):
  energy = 0
  for i in range(len(sequence)):
    letter = sequence[i]
    if(letter == 'A'):
      energy += matrix[i][0]
    if(letter == 'C'):
      energy += matrix[i][1]
    if(letter == 'G'):
      energy += matrix[i][2]
    if(letter == 'T'):
      energy += matrix[i][3]
  return energy
```

```python
"""
This code iterates through the three sequences and calls the get_energy function
defined above on each of them with the matrix determined earlier as well.
"""
sequences = ["CAGGCTCTACGCTTTATTCTTGCGGCTCGTATGGTGTGTGG", "GAGGCTGGACACTTTAATCTTCCCTATCGTATGTTGTGTGC",
             "CGAGCATGCCACCTTAAGCCTCTGGCTCGTATGACGTGTGG"]
for str in sequences:
  print(get_energy(str, energy_matrix))
```
